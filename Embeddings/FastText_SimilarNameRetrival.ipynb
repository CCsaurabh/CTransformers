```python
import pandas as pd
```

```python
df = pd.read_csv('admin.csv')

df.head()
```

```python
names = [item.lower() for item in df['Name'].tolist()]
```

```python
filename = "items_list.txt"

# Open the file in write mode ('w') to create a new file or overwrite an existing one
# Use 'a' mode to append to an existing file without overwriting it
with open(filename, 'w') as file:
    # Iterate over the list of items
    for item in names:
        # Write each item to the file followed by a newline character
        file.write(item + "\n")
```

```python
import fasttext

# Skipgram model :
model = fasttext.train_unsupervised(filename, model='skipgram')
```

```python
 model.get_nearest_neighbors('organic argan oil',k=20)
```

```python
 model.get_nearest_neighbors('citric acid',k=20,on_unicode_error='strict')
```
